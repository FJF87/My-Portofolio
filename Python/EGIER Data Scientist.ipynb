### Fungsi Polynomial

```python
# fungsi untuk menghitung nilai polynomial suatu x

def polynomial_count(x, coef):
  y = 0
  for i in range(0, len(coef)):
    if i == len(coef)-1:
      # coefisien terakhir adalah konstanta
      y += coef[i]
    else:
      y += coef[i] * x ** (len(coef)-1-i)
  return y

print(polynomial_count(2, coef_for_polynomial))
```

```python
# fungsi untuk menghitung faktorial suatu n

def factorial(n):
  result = 1
  for i in range(1, n+1):
    result *= i
  return result
```

```python
# x = angka yang ingin dicari nilainya, coef = array
# a yang digunakan adalah 0

def polynomial_taylor_version(x, coef):
  highest_order_polynomial = len(coef)-1
  a = 0
  taylor_result = 0

  coef_copy = coef.copy()
  for i in range(0, highest_order_polynomial+1):
    taylor_result += ( (polynomial_count(a, coef_copy) * (x)**(i)) / factorial(i) )
    index = 0
    coef_now = highest_order_polynomial - i
    for j in range(highest_order_polynomial+1-i):
      coef_copy[index] *= coef_now
      index += 1
      coef_now -= 1

    # konvert ndarray ke list, pop element terakhir, konvert list kembali ke ndarray
    coef_copy = list(coef_copy)
    coef_copy.pop()
    coef_copy = np.array(coef_copy)
  return taylor_result

print(polynomial_taylor_version(2, coef_for_polynomial))
```

### Fungsi Bilangan Natural

```python
x_value = 100
```

```python
def natural_func(x, alpha, beta):
  """
  natural_func(x, alpha, beta)
  fungsi untuk menghitung bilangan natural
  x = nilai x yang ingin dihitung
  alpha = nilai alpha pada persamaan bilangan natural y = alpha * e^(beta*x)
  beta = nilai beta pada persamaan bilangan natural y = alpha * e^(beta*x)
  """
  y = alpha * np.exp(beta*x)
  return y

natural_func(x_value, alpha, beta)
```

```python
def natural_taylor_version(x, alpha, beta, number_of_iteration):
  """
  natural_taylor_version(x, alpha, beta, number_of_iteration)
  a yang digunakan adalah 0
  x = nilai x yang ingin dihitung
  alpha = nilai alpha pada persamaan bilangan natural y = alpha * e^(beta*x)
  beta = nilai beta pada persamaan bilangan natural y = alpha * e^(beta*x)
  number_of_iteration = hasil yang ingin dihitung sampai turunan ke berapa
  """
  a = 0
  taylor_result = 0

  for i in range(number_of_iteration):
    taylor_result += (natural_func(a, alpha, beta) * (x)**(i)) / factorial(i)

    alpha *= beta

  return taylor_result

print(f"result = {natural_taylor_version(x_value, alpha, beta, 5)}")
```

Error taylor series dengan persamaan asli

```python
# function untuk menghitung error pada persamaan bilangan natural

def compute_error(x, alpha, beta, number_of_iteration):
  """
  compute_error(x, alpha, beta, number_of_iteration)
  x = nilai x yang ingin dihitung
  alpha = nilai alpha pada persamaan bilangan natural y = alpha * e^(beta*x)
  beta = nilai beta pada persamaan bilangan natural y = alpha * e^(beta*x)
  number_of_iteration = hasil yang ingin dihitung sampai turunan ke berapa
  """
  derivative_n = 0
  for i in range(0, number_of_iteration+1):
    alpha *= beta
  derivative_n = natural_func(x, alpha, beta)

  error = (derivative_n * (x)**(number_of_iteration+1)) / factorial(number_of_iteration+1)
  return error
```

```python
# print hasil dari function untuk menghitung error

for i in range(1, 10):
  print(f"jumlah turunan/iterasi sebanyak {i} memiliki error sebesar {compute_error(x_value, alpha, beta, i)}")
```

Dapat terlihat bahwa semakin banyak turunan/iterasi yang digunakan, maka errornya pun juga akan semakin kecil

```python
# plot data hasil perhitungan error pada setiap iterasi (hingga turunan ke berapa)

plt.figure(figsize = (12,8))
for i in range(1,10):
  plt.subplot(3,3,i)
  plt.plot(x, y)
  plt.plot(x, alpha*np.exp(beta*x), "r")
  plt.plot(x, natural_taylor_version(x, alpha, beta, i), "g")
  plt.title(f"Jumlah iterasi:{i}")

plt.tight_layout()
plt.show()
```

## Fungsi Polynomial

```python
# function untuk menghitung bisection method

def bisection_method_polynomial(x1, x2, coef, y_target, e):
  """
  bisection_method_polynomial(x1, x2, coef, y_target, e)
  x1, x2 = starting value x
  coef = list koefisien persamaan polynomial
  y_target = nilai y awal
  e = error yang diinginkan
  """
  y1 = polynomial_count(x1, coef)
  y2 = polynomial_count(x2, coef)

  if y1 > y2:
    x1, x2 = x2, x1

  if y1 < y_target and y2 < y_target:
    return "Cannot find the result because both x value are negative"
  elif y1 > y_target and y2 > y_target:
    return "Cannot find the result because both x value are positive"
  else:
    root = 0
    error = 1000
    while error > e:
      mid = (x1 + x2) / 2
      mid_value = polynomial_count(mid, coef)
      if mid_value < y_target:
        x1 = mid
      elif mid_value > y_target:
        x2 = mid
      root = mid
      error = abs(x2 - x1)
    return root
```

```python
# mencocokkan dengan memanggil library dari scipy

# ini adalah fungsi polynomial yang memiliki kekhususan, yaitu starting valuenya dibuat 0
# Hal ini disebabkan function pada scipy tidak bisa menghitung jika persaamaan awalnya tidak nol
# contoh sederhananya: y = x + 5, misal y = 2
# pada function "bisection_method_polynomial" yang saya buat di atas, saya buat agar dapat menghitung
# 2 = x + 5
# namun, pada function bawaan dari library scipy, persamaan harus diubah menjadi
# 0 = x + 3
# maka dari itulah, saya membuat fungsi spesifik ini
# karena ketika dipanggil fungsinya harus terdiri dari 1 parameter(yaitu x), maka nilai y-nya sudah saya fix-kan di dalam fungsi

def polynomial_count_spec(x):
  """
  polynomial_count_spec(x)
  function untuk menghitung polynomial secara khusus
  """
  global coef_for_polynomial
  coef = coef_for_polynomial.copy()
  coef[len(coef_for_polynomial)-1] -= 25000
  y = 0
  for i in range(0, len(coef)):
    if i == len(coef)-1:
      # coefisien terakhir adalah konstanta
      y += coef[i]
    else:
      y += coef[i] * x ** (len(coef)-1-i)
  return y

print("Untuk mencocokkan, saya memanggil dari library scipy")
import scipy as sc
print(sc.optimize.bisect(polynomial_count_spec, 100, 200))
```

## Fungsi Bilangan Natural

```python
# function untuk menghitung bisection method persamaan bilangan natural

def bisection_method_natural(x1, x2, alpha, beta, y_target, e):
  """
  bisection_method_natural(x1, x2, alpha, beta, y_target, e)

  function untuk menghitung bisection method persamaan bilangan natural

  x1, x2 = starting value x
  alpha = nilai alpha pada persamaan bilangan natural y = alpha * e^(beta*x)
  beta = nilai beta pada persamaan bilangan natural y = alpha * e^(beta*x)
  y_target = nilai y awal
  e = error yang diinginkan
  """
  y1 = natural_func(x1, alpha, beta)
  y2 = natural_func(x2, alpha, beta)

  if y1 > y2:
    x1, x2 = x2, x1

  if y1 < y_target and y2 < y_target:
    return "Cannot find the result because both x value are negative"
  elif y1 > y_target and y2 > y_target:
    return "Cannot find the result because both x value are positive"
  else:
    root = 0
    error = 1000
    while error > e:
      mid = (x1 + x2) / 2
      mid_value = natural_func(mid, alpha, beta)
      if mid_value < y_target:
        x1 = mid
      elif mid_value > y_target:
        x2 = mid
      root = mid
      error = abs(x2 - x1)
    return root
```
